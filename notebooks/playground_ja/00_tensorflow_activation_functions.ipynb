## TensorFlow の活性化関数

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

sess = tf.Session()
```

### ReLU(Rectified Linear Unit)関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.relu(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-1., 10.)
plt.grid()
plt.show()
```

### ReLU6関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.relu6(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-1., 8.)
plt.grid()
plt.show()
```

### ELU(Exponential Linear Units)関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.elu(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-2., 10.)
plt.grid()
plt.show()
```

### ソフトプラス関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.softplus(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-1., 10.)
plt.grid()
plt.show()
```

### ソフトサイン関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.softsign(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-1.0, 1.0)
plt.grid()
plt.show()
```

### シグモイド関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.sigmoid(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.5)
plt.grid()
plt.show()
```

### ハイパボリックタンジェント関数

```python
X = np.arange(-10., 10., 0.1)
Y = sess.run(tf.nn.tanh(X))

plt.plot(X, Y)
plt.xlim(-10, 10)
plt.ylim(-2., 2.)
plt.grid()
plt.show()
```
